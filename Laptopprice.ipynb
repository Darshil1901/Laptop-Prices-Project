# **Analyze the distribution of the top 10 most expensive laptops among the top 3 laptop companies.**

```python
#Top 3 Company Index
top_3_com = df['Company'].value_counts().head(3).index
top_3_com
```

```python
#New Database of the with only top 3 Company
top_3_com_df = df[df['Company'].isin(top_3_com)]
top_3_com_df
```

```python
top3_exp = top_3_com_df.groupby(['Company', 'Product'])['Price_euros'].max().sort_values(ascending=False).reset_index().head(10)
top3_exp
```

```python
plt.figure(figsize=(12,6))
sns.barplot(data=top3_exp,y ='Product', x = "Price_euros", hue = "Company", palette = "summer")
plt.title("Top 10 Expensive Laptops by Top 3 Company")
plt.ylabel("Product")
plt.xlabel("Price_euros")
plt.tight_layout()
plt.legend()
plt.show()
```

# **Overall, the analysis indicates a competitive landscape among the top three laptop companies, with Lenovo maintaining a strong position in terms of pricing and product range. HP and Dell, however, have also demonstrated their ability to offer premium laptops that can compete effectively in the market.**

# **Analyze the distribution of screen sizes, identifying key trends, patterns, and insights. Explore central tendency, dispersion, and shape to understand screen size preferences and market trends.**

```python
# Screen Size Distribution
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Inches', color = "orange", kde = True)
plt.axvline(df['Inches'].mean(), color = 'darkcyan', linestyle = 'dashed', linewidth = 2, label = f'MeanRange: {df["Inches"].mean():.2f}Inches')
plt.title('Distribution of Screen Sizes')
plt.xlabel('Screen Size (Inches)')
plt.ylabel('Frequency')
plt.tight_layout()
plt.legend()
plt.show()
```

# **The conclusion of the graph is that the distribution of screen sizes for the laptops in the dataset is skewed to the right, with a long tail extending towards larger screen sizes. The mean screen size is 15.02 inches, and the mode is 16 inches. This indicates that while the majority of laptops have screen sizes around 16 inches, there is a significant portion with larger screens, suggesting a growing demand for larger displays.**

# **Analyze the price variations among laptop models equipped with different GPU companies. Identify trends, patterns, and outliers in pricing, and explore the relationship between GPU company and product price.**

```python
gpu = df['GPU_company'].value_counts().reset_index()
gpu
```

```python
intel = df[df['GPU_company'] == 'Intel'].groupby('Product')['Price_euros'].max().sort_values(ascending=False).reset_index().head(5)
nvidia = df[df['GPU_company'] == 'Nvidia'].groupby('Product')['Price_euros'].max().sort_values(ascending=False).reset_index().head(5)
amd = df[df['GPU_company'] == 'AMD'].groupby('Product')['Price_euros'].max().sort_values(ascending=False).reset_index().head(5)
arm = df[df['GPU_company'] == 'ARM'].groupby('Product')['Price_euros'].max().sort_values(ascending=False).reset_index().head(5)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=intel['Product'], y=intel['Price_euros'], color='palegreen', label='Intel')
sns.barplot(x=nvidia['Product'], y=nvidia['Price_euros'],color='skyblue', label='Nvidia')
sns.barplot(x=amd['Product'], y=amd['Price_euros'], color='pink', label='AMD')
sns.barplot(x=arm['Product'], y=arm['Price_euros'], color='plum', label='ARM')
plt.xlabel('Product')
plt.ylabel('Price (Euros)')
plt.title('Price Distribution by GPU Company')
plt.xticks(rotation=90)
plt.legend()
plt.tight_layout()
plt.show()
```

# **The graph reveals that laptops equipped with Intel GPUs tend to be more expensive than those with Nvidia, AMD, or ARM GPUs. However, there are notable exceptions where Nvidia or AMD GPUs are utilized in higher-priced laptops, suggesting that other factors such as brand, features, or specific model configurations play a significant role in determining pricing.**
